```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/brem221/breast-cancer-prediction/main/preprocessed_breast_cancer_data.csv")
df.head(5)
```

```python
df.describe(include='all')
```

```python
sns.countplot(x='diagnosis', data=df)
```

```python
sns.pairplot(df, hue='diagnosis', vars=['radius_mean', 'texture_mean', 'area_mean'])
```

```python
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
```

```python
sns.histplot(data=df, x='radius_mean', hue='diagnosis', element='step', stat='density', common_norm=False)
```

```python
sns.violinplot(x='diagnosis', y='radius_mean', data=df, inner='quartile')
```

```python
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=df)
```

```python
diagnosis_counts = df['diagnosis'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B2FF'])
plt.title('Distribution of Diagnoses (Malignant(0) and Benign(1))')
plt.show()
```
